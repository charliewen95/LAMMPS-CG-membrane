```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
file_paths = ["r350_B1/plot_data.csv", "r350_B1.25/plot_data.csv", "r350_B1.5/plot_data.csv", "r350_B2/plot_data.csv"]
labels = ["BEND=1", "BEND=1.25", "BEND=1.5", "BEND=2"]
labels = [r"$K_{\theta} = 1.00\ k_B T\cdot \mathrm{rad}^{-2}$", 
          r"$K_{\theta} = 1.25\ k_B T\cdot \mathrm{rad}^{-2}$", 
          r"$K_{\theta} = 1.50\ k_B T\cdot \mathrm{rad}^{-2}$", 
          r"$K_{\theta} = 2.00\ k_B T\cdot \mathrm{rad}^{-2}$"]
```

```python
colors = cm.viridis(np.linspace(0, 1, len(file_paths)))
b1_data = pd.read_csv("r350_B1/plot_data.csv")
reference_y_offset = b1_data["Radius_A"].iloc[0]
```

```python
plt.figure(figsize=(10, 6))

# Loop through each file, read the data, and shift it relative to the B1 reference
for file_path, label, color in zip(file_paths, labels, colors):
    data = pd.read_csv(file_path)
    
    # Calculate the y-offset for each dataset relative to the B1 reference
    y_offset = data["Radius_A"].iloc[0] - reference_y_offset
    y_offset = 0
    
    # Shift Radius_A values to align with the first B1 value
    shifted_radius = data["Radius_A"] - y_offset
    
    # Plot the shifted data with color gradient
    plt.plot(data["Polar_angle_rad"], shifted_radius, label=label, color=color,linewidth=8)


plt.plot([0.3805188837394971,0.3805188837394971],[350,500],label='Piezo dome',color='red')
plt.plot([0.38793474014248847,0.38793474014248847],[350,500],color='red')
#plt.title("Polar Angle vs Radius (Aligned to B1) with Color Gradient")
plt.legend(fontsize=14)
plt.ylim([350,410])
plt.xlim([0,np.pi])

plt.grid()
plt.xlabel(r"Polar angle, $\theta$ [rad]",fontsize=16)
plt.ylabel("Radius [$\AA$]",fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)





plt.savefig('profiles_avg_all.png', format='png', dpi=300, bbox_inches='tight')

plt.show()
```
